```python
#1.A. Joint probability
# formula P(C)=P(A)*P(B)
#P(A)=500/2000
#P(B)=600/2000
A=500/2000
B=600/2000
C=A*B*100
print("the joint probability is (in percent)", '%.2f'% C)
```

```python
#1.B. Joint probability
# formula D=C/{P(D)}
#P(D)=500/2000
D=C/(500/2000)
print("the joint probability is (in percent)", '%.2f'% D)
```

```python
import scipy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import binom
p= 0.05
n=10
k=np.arange(0,10)
binomial = stats.binom.pmf(k,n,p)
print(binomial)
#2.A. Probability
print("2A.the probability none defective is %.2f" % binomial[0])
#2.B. Probability
print("2B.the probability that exactly one of the items is defective %.2f" % binomial[1])
cumbinomial = stats.binom.cdf(k,n,p)
print(cumbinomial)
#2.C. Probability
print("2C.the Probability that two or fewer of the items are defective %.2f" % cumbinomial[2])
#2.D. Probability
P=1-(cumbinomial[2])
print("2D.the Probability that three or more of the items are defective %.2f" % P)
```

```python
#3A.Probability that in a given week he will sell some cars. assuming in a year=52 weeks
data_poisson = poisson.rvs(mu=3, size=52)
data_poisson. mean()
print("3A Probability that in a given week he will sell some cars. assuming in a year=52 weeks is", data_poisson.mean())
```

```python
#3B Probability that in a given week he will sell 2 or more but less than 5 cars
P1=poisson.pmf(k=2, mu=3)
P2=poisson.pmf(k=3, mu=3)
P3=poisson.pmf(k=4, mu=3)
P=P1+P2+P3
print("3B Probability that in a given week he will sell 2 or more but less than 5 cars is",P)

```
